# 크롤링

```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.common.exceptions import TimeoutException
from selenium.common.exceptions import NoSuchElementException
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import Select
import time
from selenium.webdriver.common.service import Service
from bs4 import BeautifulSoup

df1 = pd.DataFrame(columns = ['이름', '행정동', '점수', '별점', '좋아요'])

dining_name = []
dining_address = []
dining_score = []
dining_userscore = []
dining_heart = []

import requests

from string import digits

chromedriver = '/Users/minjipark/Downloads/chromedriver_mac_arm64/chromedriver' 
driver = webdriver.Chrome(chromedriver)

for i, keyword in enumerate(df['동'].tolist()[200:300]):

    print("{}. {}".format(i, keyword))

    url = "https://www.diningcode.com/list.dc?query={}".format(
        keyword)
    driver.get(url)

    time.sleep(1)

    try: count = driver.find_element(By.CLASS_NAME, 'Place-Count').text
    except: count = '0'

    scroll = 0

    if int(count.replace(',', '')) >= 100 : scroll = 4
    elif int(count) % 20 == 0 : scroll = int(count) // 20 - 1
    else : scroll = int(count) // 20

    for i in range(scroll):
        driver.find_element(By.XPATH, '//button[@class="SearchMore upper"]').click()
        time.sleep(1)

    names = driver.find_elements(By.TAG_NAME, 'h2')
    scores = driver.find_elements(by=By.CLASS_NAME, value='Score')
    userScores = driver.find_elements(by=By.CLASS_NAME, value='UserScore')
    hearts = driver.find_elements(by=By.CLASS_NAME, value='heart')

    for name in names:
        dining_address.append(keyword)
        dining_name.append(name.text[3:])
    for score in scores:
        dining_score.append(score.text[:-1])
    for userScore in userScores:
        dining_userscore.append(userScore.text.split()[0])
    for heart in hearts:
        dining_heart.append(heart.text)
    
    print('{}, {}, {}, {}'.format(len(names), len(scores), len(userScores), len(hearts)))
    
driver.quit()
        
df1['이름'] = dining_name
df1['행정동'] = dining_address
df1['점수'] = dining_score
df1['별점'] = dining_userscore
df1['좋아요'] = dining_heart
```
