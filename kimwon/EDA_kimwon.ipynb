```python
expected_revenue = pd.read_csv('data/서울시_우리마을가게_상권분석서비스(신_상권_추정매출)_2021년.csv',encoding='cp949')
```

```python
expected_revenue.columns
```

```python
expected_revenue = expected_revenue[[ '상권_코드_명','서비스_업종_코드_명', '분기당_매출_금액', '분기당_매출_건수']]
```

```python
expected_revenue['mean_revenue']= expected_revenue['분기당_매출_금액']/ expected_revenue['분기당_매출_건수']
expected_revenue.drop(columns=['분기당_매출_금액','분기당_매출_건수'],inplace=True)
expected_revenue
```

```python
expected_revenue['rank_업종별'] = expected_revenue.groupby('서비스_업종_코드_명')['mean_revenue'].rank()
```

```python
expected_revenue['rank_상권별'] = expected_revenue.groupby('상권_코드_명')['mean_revenue'].rank()
```

```python
expected_revenue.drop(columns=['rank'],inplace=True)
expected_revenue
```

```python
expected_revenue.groupby('서비스_업종_코드_명')['상권_코드_명','rank_업종별'].min().to_csv('업종별 1위 상권.csv')
expected_revenue.groupby('서비스_업종_코드_명')['상권_코드_명','rank_업종별'].min()
```

```python
expected_revenue.groupby('상권_코드_명')['서비스_업종_코드_명','rank_상권별'].min().to_csv('상권별 1위 업종.csv')
expected_revenue.groupby('상권_코드_명')['서비스_업종_코드_명','rank_상권별'].min()
```
